```python
df_filtered = df[df['Curricular units 2nd sem (grade)'] != 0]


course_selection_2 = alt.selection_point(
    fields=['Course'], 
    bind=alt.binding_select(options=df_filtered['Course'].unique(), name="Select Course 2: ")  
)



course_selection_1 = alt.selection_point(
    fields=['Course'],  
    bind=alt.binding_select(options=df_filtered['Course'].unique(), name="Select Course 1: ")  
)



grade_distribution_chart_1 = alt.Chart(df_filtered).mark_bar(
    stroke = 'black',
    strokeWidth = .4
).encode(
    alt.Y('count():Q'),
    alt.X('Curricular units 2nd sem (grade):Q'),
    alt.Color('Target:N', scale=alt.Scale(domain=['Dropout', 'Graduate', 'Enrolled'], range=['#648FFF', '#CC79A7', '#117733']))  
).transform_filter(
    course_selection_1
).add_params(
    course_selection_1
).properties(
    title='Grade Distribution for Course 1', 
    width=500, 
    height=300 
)



grade_distribution_chart_2 = alt.Chart(df_filtered).mark_bar(
    stroke = 'black',
    strokeWidth = .4
).encode(
    alt.Y('count():Q'),
    alt.X('Curricular units 2nd sem (grade):Q'),
    alt.Color('Target:N', scale=alt.Scale(domain=['Dropout', 'Graduate', 'Enrolled'], range=['#648FFF', '#CC79A7', '#117733']))  
).transform_filter(
    course_selection_2
).add_params(
    course_selection_2
).properties(
    title='Grade Distribution for Course 2',
    width=500, 
    height=300 
)

   
grade_dists = grade_distribution_chart_1 | grade_distribution_chart_2

grade_dists.save('Grade_Distributions.html')
 
```

```python
#normalize data for d3 plot
df["approved_rate"] = df["Curricular units 2nd sem (approved)"] / df["Curricular units 2nd sem (enrolled)"]
df["evaluation_rate"] = df["Curricular units 2nd sem (evaluations)"] / df["Curricular units 2nd sem (enrolled)"]
df["no_eval_rate"] = df["Curricular units 2nd sem (without evaluations)"] / df["Curricular units 2nd sem (enrolled)"]

#export as json file
df_to_export = df[["Course", "approved_rate", "evaluation_rate", "no_eval_rate", "Curricular units 2nd sem (grade)"]]
df_to_export.to_json("data.json", orient="records")
```

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

#base chart
base = alt.Chart(df).mark_bar().encode(
    x=alt.X("Mother Qualification", title="Mother's Qualification"),
    y=alt.Y("Curricular units 2nd sem (grade):Q", title="Average Grade"),
    tooltip=["Mother Qualification", "Curricular units 2nd sem (grade)", "Unemployment rate", "Inflation rate", "GDP"],
)

#color gradient for unemployment rate
chart = base.encode(
    color=alt.Color("Unemployment rate:Q", scale=alt.Scale(scheme="blues"), title="Unemployment Rate")
)

#buttons to filter by mother's qualifications
qualifications_filter = alt.binding_select(options=df["Mother Qualification"].unique(), name="Select Qualification: ")
qualification_selection = alt.selection_single(fields=["Mother Qualification"], bind=qualifications_filter)

#combine chart and display
filtered_chart = chart.add_selection(qualification_selection).transform_filter(qualification_selection)

filtered_chart = filtered_chart.properties(width=600, height=400, title="Average Grade (Second Semester) by Mother's Qualifications")

filtered_chart.save('Mother_Qual_Unemployment_Grades.html')
```

```python
colorblind = ['#E69F00', '#0072B2', '#CC79A7']
selection = alt.selection_interval(encodings=['x'])

grade_scatter = alt.Chart(df).mark_circle().encode(
    alt.X('Previous qualification (grade):Q', title='Previous Qualification'),
    alt.Y('Admission grade:Q', title='Admission Grade'),
    alt.Color('Target:N', title='Dropout Status', scale=alt.Scale(domain=['Dropout', 'Graduate', 'Enrolled'], range=colorblind)),
    tooltip=['Previous qualification:N', 'Admission grade:Q', 'Target:N']
).add_params(
    selection
).properties(
    width=400, 
    height=300, 
    title='Scatterplot of Previous Qualification vs Admission Grade'
)

gender_bar = alt.Chart(df).mark_bar().encode(
    alt.X('Gender:N', title='Gender'),
    alt.Y('count():Q', title='Count'),
    alt.Color('Target:N', title='Dropout Status', scale=alt.Scale(domain=['Dropout', 'Graduate', 'Enrolled'], range=colorblind)),
).transform_filter(
    selection
).properties(
    width=400,
    height=300,
    title='Dropout Status by Gender'
)

marital_status_bar = alt.Chart(df).mark_bar().encode(
    alt.X('Marital status:N', title='Marital Status'),
    alt.Y('count():Q', title='Count'),
    alt.Color('Target:N', title='Dropout Status', scale=alt.Scale(domain=['Dropout', 'Graduate', 'Enrolled'], range=colorblind)),
).transform_filter(
    selection
).properties(
    width=400,
    height=300,
    title='Dropout Status by Marital Status'
)

scatter_and_bars = grade_scatter & (gender_bar | marital_status_bar)
scatter_and_bars.save('Qualification_Grades_Demographics.html')
```

```python
df = df[df['Curricular units 2nd sem (grade)'] != 0]
df = df[df['Curricular units 1st sem (grade)'] != 0]
df['Avg Grade'] = (df['Curricular units 1st sem (grade)'] + df['Curricular units 2nd sem (grade)']) / 2

grades_boxplot = alt.Chart(df).mark_boxplot().encode(
    alt.X('Target:N', title='Dropout Status'),  
    alt.Y('Avg Grade:Q', title='Average Grade'),  
    alt.Color('Target:N', title='Dropout Status', scale=alt.Scale(domain=['Dropout', 'Graduate', 'Enrolled'], range=colorblind)),
).properties(
    width=300, 
    height=300
).facet(
    facet=alt.Facet('Course:N', title='Average Grades by Dropout Status and Course'),
    columns=4 
)

grades_boxplot.save('Avg_Course_Grades.html')
```
